```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, LabelEncoder
```

```python
df_severidade = pd.DataFrame({
    "Severidade" : ["Baixo" , "Médio" , "Alto" , "Critico"]
})

df_severidade
```

```python
ordinal_encoder = OrdinalEncoder(categories=[["Baixo" , "Médio" , "Alto" , "Critico"]])

#Transformando os dados

df_severidade["Severidade_number"] = ordinal_encoder.fit_transform(df_severidade[["Severidade"]])

df_severidade
```

### Label Encoder

```python
pratos = pd.read_csv("pratos.csv")
pratos.head()
```

```python
label_encoding = LabelEncoder()

pratos["PRATO_RECOMENDADO_encoding"] = label_encoding.fit_transform(pratos["PRATO_RECOMENDADO"])
pratos.head(5)
```

### One-Hot Encoder

```python
pratos
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Criar encoder
one_hot_enc = OneHotEncoder(sparse_output=False)

# Ajustar e transformar
proteinas = one_hot_enc.fit_transform(pratos[["PROTEINA_ONTEM"]])

# Criar DataFrame
proteinas_df = pd.DataFrame(proteinas, columns=one_hot_enc.get_feature_names_out(["PROTEINA_ONTEM"]))

proteinas_df
```

```python
pratos_act = pd.concat([pratos, proteinas_df] , axis=1)
```
